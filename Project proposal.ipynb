## Research Question n.1:

Is there an association between a person's loneliness and the amount of physical contact they have with others?

#### Variables:

- CONNECTION_activities_hug_p3m : Categorical and ordinal. Answers the question: "In the PAST THREE MONTH, how often have you... - hugged someone?"
- CONNECTION_activities_kissed_p3m : Categorical and ordinal. Answers the question: "In the PAST THREE MONTH, how often have you... - kissed someone?"
- CONNECTION_activities_sex_p3m : Categorical and ordinal. Answers the question: "In the PAST THREE MONTH, how often have you... - had sex with someone?"
- LONELY_direct : Categorical and ordinal. Answers the question: "During the PAST WEEK, have you felt lonely?"

#### Visualizations:

For each physical contact variable (hugs: CONNECTION_activities_hug_p3m, kisses: CONNECTION_activities_kissed_p3m, and sex: CONNECTION_activities_sex_p3m), constructing a KDE will show the modality and sample size and will allow for a preliminary assesment of the variables.

#### Analysis:

The data used for analysis will be pulled from the 2021 cross-sectional, the 2022 cohort, the 2023 cross-sectional, and the 2023 cohort surveys.

A least squares linear regression model will analyze the relationship between physical contact variables and loneliness levels. This model will be constructed by assigning each physical contact variable (hugs, kisses, sex) as independent variables and LONELY_direct as the dependent variable. Bootstrapping will also be used to construct a 95% confidence interval for the slope coefficient. It is important to note here that this analysis looks to determine the relation of one physical contact variable with LONELY_direct at a time, therefore meaning the same steps will be repeated 3 times for each physical contact variable.

#### Assumptions:

Linearity, homoscedasticity, independence, and normality of residuals should hold for a least squares linear regression model.

#### Hypotheses:

Since the analysis consists of comparing three variables to a fourth, there are hypotheses for each physical contact variable.

Hugs:
- Null Hypothesis (H0): There is no association between the frequency of hugging (CONNECTION_activities_hug_p3m) and loneliness levels (LONELY_direct).
- Alternative Hypothesis (H1): There is a significant association between the frequency of hugging (CONNECTION_activities_hug_p3m) and loneliness levels (LONELY_direct).

Kisses:
- Null Hypothesis (H0): There is no association between the frequency of kissing (CONNECTION_activities_kissed_p3m) and loneliness levels (LONELY_direct).
- Alternative Hypothesis (H1): There is a significant association between the frequency of kissing (CONNECTION_activities_kissed_p3m) and loneliness levels (LONELY_direct).

Sexual Activity:
- Null Hypothesis (H0): There is no association between the frequency of sexual activity (CONNECTION_activities_sex_p3m) and loneliness levels (LONELY_direct).
- Alternative Hypothesis (H1): There is a significant association between the frequency of sexual activity (CONNECTION_activities_sex_p3m) and loneliness levels (LONELY_direct).

Since the research question looks to know whether there exists an association between a person's loneliness and the amount of physical contact they have with others, no matter the kind, another set of hypotheses should be constructed to show this,

- Null Hypothesis (H0): There is no association between any of the physical contact variables (hugs, kisses, sexual activity) and loneliness levels (LONELY_direct).
- Alternative Hypothesis (H1): At least one of the physical contact variables (hugs, kisses, sexual activity) is significantly associated with loneliness levels (LONELY_direct).

#### Possible Results:

Statistically significant negative association: The 95% CIs do not include zero and are negative.
Statistically significant positive association: The 95% CIs do not include zero and are positive.
No statistically significant association: the 95% CIs include zero.
