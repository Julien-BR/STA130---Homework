- df.shape returns just two values, the number of rows and the number of collumns, regardless of how many missing values exist in the dataset. 

- df.describe() will only analyse collumns in which all values are numeric, and not the entire dataset. Furthermore, the value it reports in the 'count' column does not include missing values, meaning the numher in count is the number of total rows in the column analysed - the number of missing values in that column.

Question 8, part 1: understanding 'df.groupby("col1")["col2"].describe()'

- df.groupby("col1"): this function takes the dataset and organises it by the different values present in column 1. If there are three different values in column one, the entire function will give you three row, if there are four different values in column 1, the function will give you 4 rows, and so on and so forth.

- ["col2"].describe: the easiest part to understant, as we've seen it already - gives the count, mean, std, min, 25%, 50%, 75%, and max of values in the second column. Important to note that "col2" stands for the name of the second column, and should be replaced with such when used.

- example of this below.

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv'
df = pd.read_csv(url)

df.groupby('survived')['pclass'].describe()
```

Question 8, part 2

- While the 'count' output of df.describe would provide a count of all non-null values for each numeric column in df, 'count' when it comes to something like df.groupby('col1')['col2'].describe would produce a count of all non null values in column 2 and only within the groups formed from column 1. 

- To better explain this, I will attempt to provide an adequate example: Say we are calculating the price, size and other unimportant values of fruits from different superstores (these are all different columns). We then apply df.groupby('col1')['col2'].'col1' has two groups, group 'Bananas' and group 'Oranges', and 'col2' contains the prices for these fruits over different stores. Now (bear with me, pretty please) each group has 20 different rows containing the values of all aformentioned columns, yet in column 2, the 'Oranges' group has 3 missing values, and the 'Bananas' group has 6 missing values. Therefore, the count of 'Oranges' will be 17 (20-3) and the count for 'Bananas' would be 14 (20-6). Thanks for sticking around for my example :D
