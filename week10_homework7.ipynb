1. Simple vs. Multiple Linear Regression:
Simple Linear Regression involves predicting an outcome based on one predictor variable, assuming a linear relationship between them.
Multiple Linear Regression uses two or more predictor variables to predict the outcome, allowing for the modeling of more complex relationships.
2. Continuous vs. Indicator Variables in Regression:
In Simple Linear Regression, a continuous predictor variable can be used directly to predict the outcome, while an indicator (dummy) variable represents categorical data with 0s and 1s to capture the presence or absence of a characteristic.
Multiple Linear Regression allows for both continuous and indicator variables as predictors. Adding an interaction term between a continuous and indicator variable allows us to model how the effect of one predictor changes depending on the level of another.
3. Interaction Effects:
Interaction between predictors occurs when the effect of one predictor on the outcome depends on the value of another predictor. For example, an interaction between TV and online advertising spend might capture how the impact of TV ads on sales changes depending on online spending (and vice versa).
Including interaction terms in regression models helps capture combined effects that might not be observed when predictors are considered independently.
4. Categorized (Binary) Variables in Regression:
When predictor variables are binary (e.g., "high" or "low" spending on TV or online ads), we use indicator variables. The regression model then estimates the effect of "high" spending relative to "low" spending, with or without interaction.
The model without interaction assumes that the effect of each advertising type is independent. The model with interaction allows the effects to depend on each other, capturing a more complex relationship (e.g., synergistic effects of combined spending).
5. Predictions from Models:
In models without interaction, predictions are based on the additive effects of predictors, assuming no synergy between them.
In models with interaction, predictions reflect how one predictor’s effect on the outcome depends on the level of the other predictor, providing a more nuanced understanding of the combined impact of multiple variables.

The apparent contradiction arises because R-squared is a measure of overall model fit (how much of the total variability is explained by the predictors), while statistical significance (p-values) tests whether a relationship exists between a specific predictor and the dependent variable, regardless of the overall model's explanatory power. In this case, while the model only explains a small portion of the variance, the individual predictors (like Generation and Sp. Def) have strong, statistically significant relationships with the outcome. However, there are likely many other unmeasured factors or complexities that the current model does not capture, which limits the overall R-squared.

moreover:
- Effect Size vs. Explained Variance: The contradiction arises because statistical significance (p-value) is about the reliability of an effect, whereas R-squared is about the amount of variance explained by the model. You can have large, statistically significant coefficients (indicating a strong relationship) but still have a low R-squared because:

    - Large Coefficients: The effect of a predictor might be large (e.g., a 20-unit difference in HP between two generations), but if there are many other factors that influence HP (which aren’t in the model), the R-squared will still be low. The model might be missing other important predictors that could explain more variability in HP.

    - Residual Variability: Even if some predictors have strong effects on HP, other random or unmeasured factors (or noise) could still explain a large portion of the remaining variability, leading to a low R-squared.

```python
import numpy as np
from sklearn.model_selection import train_test_split

fifty_fifty_split_size = int(pokeaman.shape[0]*0.5)

# Replace "NaN" (in the "Type 2" column with "None")
pokeaman.fillna('None', inplace=True)

np.random.seed(130)
pokeaman_train,pokeaman_test = \
  train_test_split(pokeaman, train_size=fifty_fifty_split_size)
pokeaman_train
```

```python
model_spec3 = smf.ols(formula='HP ~ Attack + Defense', 
                      data=pokeaman_train)
model3_fit = model_spec3.fit()
model3_fit.summary()
```

```python
yhat_model3 = model3_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model3_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model3)[0,1]**2)
```

```python
model4_linear_form = 'HP ~ Attack * Defense * Speed * Legendary'
model4_linear_form += ' * Q("Sp. Def") * Q("Sp. Atk")'
# DO NOT try adding '* C(Generation) * C(Q("Type 1")) * C(Q("Type 2"))'
# That's 6*18*19 = 6*18*19 possible interaction combinations...
# ...a huge number that will blow up your computer

model4_spec = smf.ols(formula=model4_linear_form, data=pokeaman_train)
model4_fit = model4_spec.fit()
model4_fit.summary()
```

```python
yhat_model4 = model4_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model4_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model4)[0,1]**2)
```

```python
# Here's something a little more reasonable...
model6_linear_form = 'HP ~ Attack + Speed + Q("Sp. Def") + Q("Sp. Atk")'
# And here we'll add the significant indicators from the previous model
# https://chatgpt.com/share/81ab88df-4f07-49f9-a44a-de0cfd89c67c
model6_linear_form += ' + I(Q("Type 1")=="Normal")'
model6_linear_form += ' + I(Q("Type 1")=="Water")'
model6_linear_form += ' + I(Generation==2)'
model6_linear_form += ' + I(Generation==5)'

model6_spec = smf.ols(formula=model6_linear_form, data=pokeaman_train)
model6_fit = model6_spec.fit()
model6_fit.summary()

yhat_model6 = model6_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model6_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model6)[0,1]**2)
```

```python
# And here's a slight change that seems to perhaps improve prediction...
model7_linear_form = 'HP ~ Attack * Speed * Q("Sp. Def") * Q("Sp. Atk")'
model7_linear_form += ' + I(Q("Type 1")=="Normal")'
model7_linear_form += ' + I(Q("Type 1")=="Water")'
model7_linear_form += ' + I(Generation==2)'
model7_linear_form += ' + I(Generation==5)'

model7_spec = smf.ols(formula=model7_linear_form, data=pokeaman_train)
model7_fit = model7_spec.fit()
model7_fit.summary()

yhat_model7 = model7_fit.predict(pokeaman_test)
y = pokeaman_test.HP
print("'In sample' R-squared:    ", model7_fit.rsquared)
print("'Out of sample' R-squared:", np.corrcoef(y,yhat_model7)[0,1]**2)

# And here's a slight change that seems to perhas improve prediction...
model7_linear_form_CS = 'HP ~ scale(center(Attack)) * scale(center(Speed))'
model7_linear_form_CS += ' * scale(center(Q("Sp. Def"))) * scale(center(Q("Sp. Atk")))'
# We DO NOT center and scale indicator variables
model7_linear_form_CS += ' + I(Q("Type 1")=="Normal")'
model7_linear_form_CS += ' + I(Q("Type 1")=="Water")'
model7_linear_form_CS += ' + I(Generation==2)'
model7_linear_form_CS += ' + I(Generation==5)'

model7_CS_spec = smf.ols(formula=model7_linear_form_CS, data=pokeaman_train)
model7_CS_fit = model7_CS_spec.fit()
model7_CS_fit.summary().tables[-1] 
# "Cond. No." is NOW 15.4 due to centering and scaling
```
